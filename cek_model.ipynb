```python
import pickle
import pandas as pd

# Ganti path sesuai lokasi file model.pkl kamu
model_path = "model.pkl"

# Load model
with open(model_path, "rb") as f:
    model = pickle.load(f)

# Tampilkan tipe dan struktur model
print("=== Struktur Model ===")
print(model)
print("\n")

# Coba lihat fitur yang diharapkan model (jika tersedia)
if hasattr(model, "feature_names_in_"):
    print("Fitur yang digunakan model:")
    print(model.feature_names_in_)
else:
    print("⚠️ Model tidak memiliki atribut feature_names_in_")
    print("Mungkin bukan pipeline lengkap (perlu preprocessing manual)")

# Buat data contoh untuk uji coba prediksi
sample = pd.DataFrame([{
    "model": "Audi A1",
    "year": 2017,
    "transmission": "Manual",
    "mileage": 35000,
    "fuelType": "Petrol",
    "tax": 30,
    "mpg": 55.4,
    "engineSize": 1.4
}])

print("\n=== Contoh Input ===")
print(sample)

# Uji prediksi
try:
    pred = model.predict(sample)
    print("\n✅ Prediksi berhasil:")
    print(pred)
except Exception as e:
    print("\n❌ Gagal melakukan prediksi:")
    print(e)

```
